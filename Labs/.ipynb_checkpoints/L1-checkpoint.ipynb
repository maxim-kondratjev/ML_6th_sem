```python
def null_counter(data): # Количество пустых значений в колонке
    for col in data.columns:
        temp_null_count = data[data[col].isnull()].shape[0]
        print('{} - {}'.format(col, temp_null_count))       
    print('\n')
```

```python
null_counter(players)
null_counter(p_details)
null_counter(stats)
```

```python
def nul_deletion(data, coef):
    for col in data.columns:
        temp_null_count = data[data[col].isnull()].shape[0]
        if temp_null_count>data[col].shape[0]*coef:
            del data[col]
        else:
            if data[col].dtype != 'O':
                data[col]=data[col].fillna(data[col].median())
            data.fillna(method='bfill', inplace =True)
    data.fillna(method='ffill', inplace =True)
```

```python
nul_deletion(players, 0.3)
nul_deletion(stats, 0.3)
nul_deletion(p_details, 0.3)
```

```python
null_counter(players)
null_counter(p_details)
null_counter(stats)
```

```python
players.dropna(subset=['Player'], how='all', inplace = True)
stats.dropna(subset=['Player'], how='all', inplace = True)
```

```python
null_counter(players)
null_counter(p_details)
null_counter(stats)
```
