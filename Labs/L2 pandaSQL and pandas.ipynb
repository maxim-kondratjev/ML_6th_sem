```python
def null_counter(data): # Количество пустых значений в колонке
    for col in data.columns:
        temp_null_count = data[data[col].isnull()].shape[0]
        print('{} - {}'.format(col, temp_null_count))       
    print('\n')
```

```python
def aggr_pandasql(data):
    aggr_query = '''
        SELECT
            min(Player_id) as Player_id,
            min(Player) as Player,
            avg(G) as G,
            avg(PF) as PF,
            avg(PTS) as PTS
        FROM data 
        GROUP BY Player
        ORDER BY Player_id
        '''
    return ps.sqldf(aggr_query, locals())
```

```python
%%time
aggr_stats = aggr_pandasql(SQLstats)
```

```python
aggr_stats = aggr_stats.reset_index(drop=True)
for i in range(aggr_stats.shape[0]):
    aggr_stats.at[i, 'Player_id'] = i
aggr_stats.head()
```

```python
def join_pandasql(stats, players):
    join_query = '''
        SELECT
            *
            FROM stats
            INNER JOIN players
            on stats.Player_id = players.PLayer_id
            '''
    return ps.sqldf(join_query, locals())
```
